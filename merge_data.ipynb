```python
import pandas as pd
import glob
import os

path = "data_csv"

# Get all CSV files inside the folder
all_files = glob.glob(os.path.join(path, "*.csv"))

# Read and merge
dfs = [pd.read_csv(f) for f in all_files]
merged_df = pd.concat(dfs, ignore_index=True)
```
